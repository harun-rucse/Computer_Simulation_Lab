```python
import matplotlib.pyplot as plt
import numpy as np

#Initial values
dt = 0.1
k1 = 0.008
k2 = 0.002
N = 100

A = [0]*N
B = [0]*N
C = [0]*N

'''   ---- chemical reaction----
        A       +     B  =      C
       1 mole       1 mole      2 mole
      100 gm        50 gm      0 gm     (initial state)

'''

A[0] = 100
B[0] = 50
C[0] = 0

# Compute chemical reactor and product
for i in range(N-1):
    d_a = (k2*C[i] - k1*A[i]*B[i])*dt
    d_b = (k2*C[i] - k1*A[i]*B[i])*dt
    d_c = 2*(k1*A[i]*B[i] - k2*C[i])*dt
    
    A[i+1] = A[i] + d_a
    B[i+1] = B[i] + d_b
    C[i+1] = C[i] + d_c
    
# Print result
T = 0
print('TIME      A(i)         B(i)        C(i)')
print('----------------------------------------')

for i in range(N):
    print("{:.2f}      {:.2f}        {:.2f}        {:.2f}".format(T, A[i], B[i], C[i]))
    T = T + dt
    
# Graph
x = np.arange(0, dt*N, dt)
plt.plot(x, A, label="reactant A", color="green")
plt.plot(x, B, label="reactant B", color="blue")
plt.plot(x, C, label="reactant C", color="red")
plt.legend()
plt.show()
    
```
